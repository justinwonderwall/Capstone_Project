## Modelling

I have tried 3 modelling techniques , we will use the best one of these 3

### Logistic Regression

```python

#  Logistic Regression — Classifying Unusual Funding Records


from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay


# 1. Define Anomaly Threshold using IQR Rule


Q1 = df['CSP_commonwealth_contribution'].quantile(0.25)
Q3 = df['CSP_commonwealth_contribution'].quantile(0.75)
IQR = Q3 - Q1
upper_threshold = Q3 + 1.5 * IQR

# Label anomalies
df['anomaly_label'] = np.where(df['CSP_commonwealth_contribution'] > upper_threshold, 1, 0)

print(f"Upper threshold for anomaly: {upper_threshold:.2f}")
print(df['anomaly_label'].value_counts())


# 2. Select Features for Classification


features = ['eftsl_2024',
            'max_student_contrib_2024',
            'commonwealth_contrib_2024',
            'funding_cluster',
            'course_type_broad',
            'foe_broad']

target = 'anomaly_label'

X = df[features]
y = df[target]


# 3. Split Data into Train/Test


X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)


# 4.Preprocessing — Scale Numerics + Encode Categoricals


num_features = ['eftsl_2024', 'max_student_contrib_2024', 'commonwealth_contrib_2024']
cat_features = ['funding_cluster', 'course_type_broad', 'foe_broad']

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num_features),
        ('cat', OneHotEncoder(drop='first'), cat_features)
    ])


# 5. Logistic Regression Pipeline


log_reg = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression(max_iter=1000, class_weight='balanced', random_state=42))
])

# Train model
log_reg.fit(X_train, y_train)

# Predict
y_pred = log_reg.predict(X_test)


# 6. Evaluation


print(" Classification Report:\n", classification_report(y_test, y_pred))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Normal', 'High Funding'])
disp.plot(cmap='Blues')
plt.title("Logistic Regression — Confusion Matrix")
plt.show()


# 7. Feature Importance (from coefficients)


# Get model coefficients
coefficients = log_reg.named_steps['classifier'].coef_[0]
feature_names = list(preprocessor.named_transformers_['num'].get_feature_names_out(num_features)) + \
                list(preprocessor.named_transformers_['cat'].get_feature_names_out(cat_features))

importance_df = pd.DataFrame({
    'Feature': feature_names,
    'Coefficient': coefficients,
    'AbsValue': np.abs(coefficients)
}).sort_values(by='AbsValue', ascending=False)

# Display
print("\nFeature Importance (Top 10):")
display(importance_df.head(10))

# Plot
plt.figure(figsize=(8,5))
sns.barplot(data=importance_df.head(10), x='AbsValue', y='Feature', palette='viridis')
plt.title('Top 10 Most Influential Features (Logistic Regression)')
plt.xlabel('|Coefficient| (Impact Strength)')
plt.ylabel('Feature')
plt.show()
```

### Random Forest Classifier

```python

#  RANDOM FOREST MODEL — Predicting High Funding (Anomaly) Cases


from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Define features and target
features = ['eftsl_2024',
            'max_student_contrib_2024',
            'commonwealth_contrib_2024',
            'funding_cluster',
            'course_type_broad',
            'foe_broad']
target = 'anomaly_label'

X = df[features]
y = df[target]

# Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# Preprocessor
num_features = ['eftsl_2024', 'max_student_contrib_2024', 'commonwealth_contrib_2024']
cat_features = ['funding_cluster', 'course_type_broad', 'foe_broad']

preprocessor = ColumnTransformer([
    ('num', StandardScaler(), num_features),
    ('cat', OneHotEncoder(drop='first'), cat_features)
])

# Random Forest Pipeline
rf_model = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(
        n_estimators=200,
        max_depth=10,
        class_weight='balanced',
        random_state=42
    ))
])

# Train
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_test)

# Evaluate
print(" Random Forest Classification Report:\n", classification_report(y_test, y_pred_rf))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred_rf)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Normal', 'High Funding'])
disp.plot(cmap='Greens')
plt.title("Random Forest — Confusion Matrix")
plt.show()

# Feature Importance
feature_names = list(preprocessor.named_transformers_['num'].get_feature_names_out(num_features)) + \
                list(preprocessor.named_transformers_['cat'].get_feature_names_out(cat_features))

rf_importance = rf_model.named_steps['classifier'].feature_importances_
rf_imp_df = pd.DataFrame({'Feature': feature_names, 'Importance': rf_importance}).sort_values(by='Importance', ascending=False)

print("\nTop 10 Important Features:")
display(rf_imp_df.head(10))

plt.figure(figsize=(8,5))
sns.barplot(data=rf_imp_df.head(10), x='Importance', y='Feature', palette='crest')
plt.title('Random Forest — Top 10 Feature Importances')
plt.show()
```

### XGBoost Classifier

```python

#  XGBOOST MODEL — Predicting High Funding (Anomaly) Cases


from xgboost import XGBClassifier
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay

# XGB Pipeline
xgb_model = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', XGBClassifier(
        n_estimators=300,
        learning_rate=0.1,
        max_depth=6,
        subsample=0.8,
        colsample_bytree=0.8,
        scale_pos_weight=float(len(y[y==0]) / len(y[y==1])),
        random_state=42,
        use_label_encoder=False,
        eval_metric='logloss'
    ))
])

# Train
xgb_model.fit(X_train, y_train)
y_pred_xgb = xgb_model.predict(X_test)

# Evaluate
print(" XGBoost Classification Report:\n", classification_report(y_test, y_pred_xgb))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred_xgb)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Normal', 'High Funding'])
disp.plot(cmap='Oranges')
plt.title("XGBoost — Confusion Matrix")
plt.show()

# Feature Importance
feature_names = list(preprocessor.named_transformers_['num'].get_feature_names_out(num_features)) + \
                list(preprocessor.named_transformers_['cat'].get_feature_names_out(cat_features))

xgb_importance = xgb_model.named_steps['classifier'].feature_importances_
xgb_imp_df = pd.DataFrame({'Feature': feature_names, 'Importance': xgb_importance}).sort_values(by='Importance', ascending=False)

print("\nTop 10 Important Features:")
display(xgb_imp_df.head(10))

plt.figure(figsize=(8,5))
sns.barplot(data=xgb_imp_df.head(10), x='Importance', y='Feature', palette='rocket')
plt.title('XGBoost — Top 10 Feature Importances')
plt.show()
```

### Model Comparison (Logistic vs RF vs XGBoost)

```python

#  MODEL COMPARISON — Logistic vs Random Forest vs XGBoost


from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, roc_curve
import numpy as np

# Get probabilities for ROC curves
y_proba_log = log_reg.predict_proba(X_test)[:,1]
y_proba_rf  = rf_model.predict_proba(X_test)[:,1]
y_proba_xgb = xgb_model.predict_proba(X_test)[:,1]

# Evaluation Function
def evaluate_model(name, y_true, y_pred, y_proba):
    return {
        'Model': name,
        'Accuracy': accuracy_score(y_true, y_pred),
        'Precision': precision_score(y_true, y_pred),
        'Recall': recall_score(y_true, y_pred),
        'F1': f1_score(y_true, y_pred),
        'ROC_AUC': roc_auc_score(y_true, y_proba)
    }

# Predictions
y_pred_log = log_reg.predict(X_test)

# Collect results
results = []
results.append(evaluate_model("Logistic Regression", y_test, y_pred_log, y_proba_log))
results.append(evaluate_model("Random Forest", y_test, y_pred_rf, y_proba_rf))
results.append(evaluate_model("XGBoost", y_test, y_pred_xgb, y_proba_xgb))

results_df = pd.DataFrame(results).set_index('Model')
display(results_df.style.background_gradient(cmap='Blues').format("{:.3f}"))


#  ROC CURVE COMPARISON

plt.figure(figsize=(8,6))
for name, y_proba in zip(
    ['Logistic Regression', 'Random Forest', 'XGBoost'],
    [y_proba_log, y_proba_rf, y_proba_xgb]
):
    fpr, tpr, _ = roc_curve(y_test, y_proba)
    plt.plot(fpr, tpr, label=f"{name} (AUC={roc_auc_score(y_test, y_proba):.3f})")

plt.plot([0,1], [0,1], 'k--')
plt.title("ROC Curve Comparison — Logistic vs Random Forest vs XGBoost")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.grid(True)
plt.show()


#  Summary of Best Model


best_model = results_df['ROC_AUC'].idxmax()
best_auc = results_df.loc[best_model, 'ROC_AUC']
print(f" Best Model: {best_model} with ROC-AUC = {best_auc:.3f}")

if best_model == 'Random Forest':
    print(" Random Forest performed best — captures non-linear patterns between EFTSL and contributions.")
elif best_model == 'XGBoost':
    print(" XGBoost performed best — handles rare anomalies and complex interactions effectively.")
else:
    print("ℹ Logistic Regression performed best — data likely has mostly linear relationships.")
```

## Unsupervised Modelling

### Goal

Find natural groupings among Course × Unit × FOE combinations based on:

* Funding and load characteristics (EFTSL, contributions, etc.)

* Discipline-level patterns (Funding Cluster, FOE)

This helps to answer:

“Which types of course-unit combinations behave similarly in funding and study load — and which ones stand apart?”

```python

#  UNSUPERVISED LEARNING — K-MEANS CLUSTERING


from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# --Select numeric features for clustering ---
features_cluster = [
    'eftsl_2024',
    'max_student_contrib_2024',
    'commonwealth_contrib_2024',
    'CSP_student_contribution',
    'CSP_commonwealth_contribution'
]

X_cluster = df[features_cluster].copy().dropna()

# --- Scale features ---
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_cluster)

# --- Find optimal number of clusters (Elbow & Silhouette) ---
inertia = []
silhouette = []
k_range = range(2, 10)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)
    silhouette.append(silhouette_score(X_scaled, kmeans.labels_))

# --- Elbow Method Plot ---
plt.figure(figsize=(7,5))
plt.plot(k_range, inertia, 'o-', color='#1d3557', linewidth=2)
plt.title("Elbow Method — Optimal Number of Clusters", fontsize=14, fontweight='bold')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia (Within-Cluster Sum of Squares)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()


```

```python
# --- Silhouette Score Plot ---
plt.figure(figsize=(7,5))
plt.plot(k_range, silhouette, 'o-', color='#e63946', linewidth=2)
plt.title("Silhouette Scores for Different k", fontsize=14, fontweight='bold')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

```

```python
# ---  Choose best k based on Elbow & Silhouette ---
best_k = k_range[np.argmax(silhouette)]
print(f"✅ Optimal number of clusters (based on silhouette): {best_k}")

# --- Fit final K-Means with best_k ---
kmeans_final = KMeans(n_clusters=best_k, random_state=42, n_init=10)
df['cluster_kmeans'] = kmeans_final.fit_predict(X_scaled)

# ---  Summary of cluster means ---
cluster_summary = df.groupby('cluster_kmeans')[features_cluster].mean().round(2)
print("\nCluster Mean Profiles:")
display(cluster_summary)

# ---  Visualize clusters (pairwise funding relationships) ---
plt.figure(figsize=(8,6))
sns.scatterplot(
    x='eftsl_2024',
    y='CSP_commonwealth_contribution',
    hue='cluster_kmeans',
    palette='Set2',
    data=df,
    alpha=0.7,
    s=60
)
plt.title("K-Means Clustering — EFTSL vs Commonwealth Contribution", fontsize=14, fontweight='bold')
plt.xlabel("EFTSL (Equivalent Full-Time Student Load)")
plt.ylabel("Commonwealth Contribution ($)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title="Cluster", fontsize=10)
plt.show()
```

The modelling process combined both supervised and unsupervised approaches to uncover funding anomalies and structural patterns within dataset

###  Supervised Modelling (Anomaly Detection):
Three classification algorithms were compared — Logistic Regression, Random Forest, and XGBoost.
XGBoost achieved the best overall performance with an ROC-AUC of 0.972 and strong precision–recall balance, confirming that ensemble-based tree methods capture the non-linear relationships between EFTSL, student contribution, and Commonwealth contribution.
The model successfully flagged high-funding outliers, particularly records with unusually large government payments relative to EFTSL.
Visual analyses (ROC, precision–recall, and funding scatter plots) indicated that most courses follow expected load-to-funding trends, while a small subset deviates sharply — potential cases for audit or policy review.

### Unsupervised Modelling (Clustering):
K-Means clustering, guided by silhouette analysis, identified eight distinct funding profiles.
Each cluster represented a unique combination of load and funding sources — for example:

Cluster 4: High-EFTSL programs with strong Commonwealth support (e.g., STEM and Health).

Cluster 5: Moderate load but high student contribution (likely self-funded or professional courses).

Cluster 1: Low total funding and small EFTSL, suggesting short units or niche offerings.
Principal Component Analysis (PCA) confirmed meaningful separation between these clusters, with the first two components explaining roughly 70–80 % of the total variance.

### Interpretation:
The clustering outcome highlights the internal structure of UWA’s funding ecosystem — from government-heavy programs to student-funded units — offering insight into where resources are concentrated or potentially mis-classified.
By linking anomaly scores with cluster membership, the analysis provides a practical monitoring framework: unusual entries can be contextualised within their peer cluster to decide whether the variance reflects legitimate diversity or a reporting anomaly.
